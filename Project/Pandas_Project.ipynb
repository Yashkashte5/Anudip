***Importing Libraries***

```python
import pandas as pd
import matplotlib.pyplot as plt
```

***Importing the Dataset used***

```python
data_df=pd.read_csv("/content/netflix_titles.csv")
```

```python
data_df
```

```python
data_df.columns
```

***Displaying count of type and plotting it***

```python
data_df['type'].value_counts()
data_df['type'].value_counts().plot(kind='bar')
```

***Top 10 Directors with the Most Movies and TV Shows on Netflix***

```python

director_counts = data_df['director'].value_counts()
top_directors = director_counts.head(10)
plt.figure(figsize=(10, 6))
top_directors.plot(kind='bar')

plt.xlabel('Director')
plt.ylabel('Number of Movies and TV Shows')
plt.title('Top 10 Directors with the Most Movies and TV Shows on Netflix')

plt.xticks(rotation=45)

plt.show()
```

***Number of Movies and TV Shows Released on Netflix from 2000 to 2021***

```python

filtered_df = data_df[data_df["release_year"] >= 2000]

grouped_df = filtered_df.groupby("release_year").size()

grouped_df.plot(kind="bar")

plt.xlabel("Release Year")
plt.ylabel("Number of Movies and TV Shows")
plt.title("Number of Movies and TV Shows Released on Netflix from 2000 to 2021")

plt.show()
```

***Listing the ratings and their count accordingly***

```python

ratings_counts = data_df['rating'].value_counts()
print(ratings_counts)
```

***Displaying the top 10 directors***

```python
top_directors = director_counts.head(10)

print("Top 10 Directors with the Most Movies and TV Shows on Netflix:")
print(top_directors)
```

***Top 10 Countries with the Most Movies and TV Shows on Netflix***

```python

country_counts = data_df['country'].value_counts()
top_countries = country_counts.head(10)
top_countries.plot(kind='bar')

plt.xlabel('Country')
plt.ylabel('Number of Movies and TV Shows')
plt.title('Top 10 Countries with the Most Movies and TV Shows on Netflix')

plt.xticks(rotation=45)

plt.show()
```

***List of Genres***
